```python
# Named Entity Recognition

import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer, word_tokenize


training_text = state_union.raw("1993-Clinton.txt")
testing_text = state_union.raw("2000-Clinton.txt")

# Train the Tokenizer
sent_tokenizer = PunktSentenceTokenizer(training_text)

# Tokenize the Sentences; converts text to list of sentences;(sent_tokenize)
tokenized = sent_tokenizer.tokenize(testing_text)
#print('Tokenized sentences: \n\n',tokenized)

print()

def process_text():
    try:
        for word in tokenized:
            words = word_tokenize(word)
            tagged = nltk.pos_tag(words)
            namedEnt = nltk.ne_chunk(tagged, binary=True)
            #namedEnt.draw()
            print(namedEnt)

    except Exception as e:
        print(str(e))

process_text()
```
